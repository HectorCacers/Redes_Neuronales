```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import json

class MNISTTrainer:
    def __init__(self):
        self.train_images = None
        self.train_labels = None
        self.test_images = None
        self.test_labels = None
        
        # Hiperparámetros (ajustables)
        self.input_size = 784    # 28x28
        self.hidden_size = 128   # neuronas capa oculta
        self.output_size = 10    # dígitos 0-9
        self.learning_rate = 0.01
        self.epochs = 10
        self.batch_size = 64
        
        # Pesos y sesgos
        self.W1 = None
        self.b1 = None
        self.W2 = None
        self.b2 = None
    
    def load_data(self):
        """Cargar datos MNIST"""
        # Datos de entrenamiento
        train_data = np.load("mnist_train.npz")
        self.train_images = train_data["images"]
        self.train_labels = train_data["labels"]
        
        # Datos de prueba
        test_data = np.load("mnist_test.npz")
        self.test_images = test_data["images"]
        self.test_labels = test_data["labels"]
        
        print(" Datos cargados:")
        print(f"   Entrenamiento: {self.train_images.shape} imágenes")
        print(f"   Prueba: {self.test_images.shape} imágenes")
    
    def preprocess_data(self):
        """Preprocesar datos"""
        # Normalizar imágenes (0-255 -> 0-1)
        self.X_train = self.train_images.reshape(-1, 784) / 255.0
        self.X_test = self.test_images.reshape(-1, 784) / 255.0
        
        # Convertir labels a one-hot encoding
        self.y_train = np.eye(10)[self.train_labels]
        self.y_test = self.test_labels
        
        print(" Datos preprocesados")
    
    def initialize_weights(self):
        """Inicializar pesos y sesgos"""
        # Capa oculta
        self.W1 = np.random.randn(self.input_size, self.hidden_size) * 0.01
        self.b1 = np.zeros((1, self.hidden_size))
        
        # Capa de salida
        self.W2 = np.random.randn(self.hidden_size, self.output_size) * 0.01
        self.b2 = np.zeros((1, self.output_size))
        
        print(" Pesos inicializados")
    
    def relu(self, x):
        """Función de activación ReLU"""
        return np.maximum(0, x)
    
    def relu_derivative(self, x):
        """Derivada de ReLU"""
        return (x > 0).astype(float)
    
    def softmax(self, x):
        """Función Softmax"""
        exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
        return exp_x / np.sum(exp_x, axis=1, keepdims=True)
    
    def forward_propagation(self, X):
        """Propagación hacia adelante"""
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.relu(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.softmax(self.z2)
        return self.a2
    
    def backward_propagation(self, X, y, output):
        """Propagación hacia atrás"""
        m = X.shape[0]
        
        # Capa de salida
        dz2 = output - y
        dW2 = (1/m) * np.dot(self.a1.T, dz2)
        db2 = (1/m) * np.sum(dz2, axis=0, keepdims=True)
        
        # Capa oculta
        dz1 = np.dot(dz2, self.W2.T) * self.relu_derivative(self.z1)
        dW1 = (1/m) * np.dot(X.T, dz1)
        db1 = (1/m) * np.sum(dz1, axis=0, keepdims=True)
        
        return dW1, db1, dW2, db2
    
    def update_parameters(self, dW1, db1, dW2, db2):
        """Actualizar pesos y sesgos"""
        self.W1 -= self.learning_rate * dW1
        self.b1 -= self.learning_rate * db1
        self.W2 -= self.learning_rate * dW2
        self.b2 -= self.learning_rate * db2
    
    def compute_loss(self, y_true, y_pred):
        """Calcular pérdida"""
        m = y_true.shape[0]
        loss = -np.sum(y_true * np.log(y_pred + 1e-8)) / m
        return loss
    
    def train(self):
        """Entrenar el modelo"""
        print("🚀 Iniciando entrenamiento...")
        print(f"   Épocas: {self.epochs}")
        print(f"   Tasa de aprendizaje: {self.learning_rate}")
        print(f"   Batch size: {self.batch_size}")
        
        losses = []
        accuracies = []
        
        for epoch in range(self.epochs):
            epoch_loss = 0
            num_batches = len(self.X_train) // self.batch_size
            
            for i in range(num_batches):
                # Mini-batch
                start = i * self.batch_size
                end = start + self.batch_size
                X_batch = self.X_train[start:end]
                y_batch = self.y_train[start:end]
                
                # Forward propagation
                output = self.forward_propagation(X_batch)
                
                # Calcular pérdida
                loss = self.compute_loss(y_batch, output)
                epoch_loss += loss
                
                # Backward propagation
                dW1, db1, dW2, db2 = self.backward_propagation(X_batch, y_batch, output)
                
                # Actualizar parámetros
                self.update_parameters(dW1, db1, dW2, db2)
            
            # Calcular precisión después de cada época
            train_acc = self.evaluate_accuracy(self.X_train, self.train_labels)
            test_acc = self.evaluate_accuracy(self.X_test, self.test_labels)
            
            avg_loss = epoch_loss / num_batches
            losses.append(avg_loss)
            accuracies.append(test_acc)
            
            print(f"📊 Época {epoch+1}/{self.epochs} - Loss: {avg_loss:.4f} - Train Acc: {train_acc:.4f} - Test Acc: {test_acc:.4f}")
        
        return losses, accuracies
    
    def evaluate_accuracy(self, X, y):
        """Evaluar precisión"""
        predictions = self.predict(X)
        return accuracy_score(y, predictions)
    
    def predict(self, X):
        """Hacer predicciones"""
        output = self.forward_propagation(X)
        return np.argmax(output, axis=1)
    
    def save_model(self, filename):
        """Guardar modelo en formato JSON"""
        model_config = {
            "input_shape": [28, 28, 1],
            "preprocess": {"scale": 255.0},
            "layers": [
                {
                    "units": self.hidden_size,
                    "activation": "relu",
                    "W": self.W1.tolist(),
                    "b": self.b1.tolist()
                },
                {
                    "units": self.output_size,
                    "activation": "softmax", 
                    "W": self.W2.tolist(),
                    "b": self.b2.tolist()
                }
            ]
        }
        
        with open(filename, 'w') as f:
            json.dump(model_config, f, indent=2)
        
        print(f"✅ Modelo guardado como {filename}")
    
    def plot_training_history(self, losses, accuracies):
        """Graficar historial de entrenamiento"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Gráfica de pérdida
        ax1.plot(losses)
        ax1.set_title('Pérdida durante entrenamiento')
        ax1.set_xlabel('Época')
        ax1.set_ylabel('Loss')
        
        # Gráfica de precisión
        ax2.plot(accuracies)
        ax2.set_title('Precisión durante entrenamiento')
        ax2.set_xlabel('Época')
        ax2.set_ylabel('Accuracy')
        
        plt.tight_layout()
        plt.show()

# Ejecutar entrenamiento completo
def main():
    print(" ENTRENAMIENTO DE MODELO DESDE CERO")
    print("=" * 50)
    
    # Crear y entrenar modelo
    trainer = MNISTTrainer()
    
    # 1. Cargar datos
    trainer.load_data()
    
    # 2. Preprocesar
    trainer.preprocess_data()
    
    # 3. Inicializar pesos
    trainer.initialize_weights()
    
    # 4. Entrenar modelo
    losses, accuracies = trainer.train()
    
    # 5. Evaluar precisión final
    final_accuracy = trainer.evaluate_accuracy(trainer.X_test, trainer.test_labels)
    print(f"\n PRECISIÓN FINAL DEL MODELO: {final_accuracy:.4f} ({final_accuracy*100:.2f}%)")
    
    # 6. Guardar modelo
    trainer.save_model("mi_modelo_entrenado.json")
    
    # 7. Mostrar gráficas
    trainer.plot_training_history(losses, accuracies)
    
    print("\n Entrenamiento completado exitosamente!")

# Ejecutar
if __name__ == "__main__":
    main()
```
