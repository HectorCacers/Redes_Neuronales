```python
import numpy as np
import json

class MNISTClassifier:
    def __init__(self):
        self.weights = []
        self.biases = []
    
    def load_model(self, model_file):
        """Cargar el modelo desde el archivo JSON"""
        with open(model_file, 'r') as f:
            model_config = json.load(f)
        
        # Extraer pesos y sesgos de cada capa
        for layer in model_config['layers']:
            if 'W' in layer and 'b' in layer:
                self.weights.append(np.array(layer['W']))
                self.biases.append(np.array(layer['b']))
        
        print(" Modelo cargado exitosamente")
    
    def load_test_data(self, test_file="mnist_test.npz"):
        """Cargar datos de prueba MNIST"""
        test_data = np.load(test_file)
        self.test_images = test_data["images"]
        self.test_labels = test_data["labels"]
        print(f" Datos de prueba cargados: {self.test_images.shape}")
    
    def preprocess_image(self, image):
        """Preprocesar una imagen"""
        normalized = image.astype(np.float32) / 255.0
        return normalized.reshape(-1)
    
    def relu(self, x):
        return np.maximum(0, x)
    
    def softmax(self, x):
        exp_x = np.exp(x - np.max(x))
        return exp_x / np.sum(exp_x)
    
    def predict_digit(self, image):
        """Predecir dígito de una imagen"""
        x = self.preprocess_image(image)
        
        for i, (W, b) in enumerate(zip(self.weights, self.biases)):
            x = np.dot(x, W) + b
            if i < len(self.weights) - 1:  # ReLU para capas ocultas
                x = self.relu(x)
            else:  # Softmax para última capa
                x = self.softmax(x)
        
        return np.argmax(x)
    
    def evaluate_accuracy(self, num_samples=10000):
        """Evaluar precisión del modelo"""
        print(f"\n Evaluando modelo con {num_samples} muestras de prueba...")
        
        correct = 0
        
        indices = range(min(num_samples, len(self.test_images)))
        
        for idx in indices:
            image = self.test_images[idx]
            true_label = self.test_labels[idx]
            predicted_digit = self.predict_digit(image)
            
            if predicted_digit == true_label:
                correct += 1
        
        accuracy = correct / len(indices)
        print(f" PRECISIÓN FINAL: {accuracy:.4f} ({accuracy*100:.2f}%)")
        print(f" Predicciones correctas: {correct}/{len(indices)}")
        
        return accuracy


print(" Evaluación de Precisión del Modelo MNIST")
print("=" * 50)


classifier = MNISTClassifier()


classifier.load_model("mnist_mlp_pretty.json")


classifier.load_test_data("mnist_test.npz")


accuracy = classifier.evaluate_accuracy(10000)  

print(f"\n EVALUACIÓN COMPLETADA - Precisión: {accuracy*100:.2f}%")
```
